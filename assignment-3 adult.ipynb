```python
def salary_data(sal):
    if sal=='<=50k':
        return 0
    else:
        return 1

data['new_encoded_sal']=data['income'].apply(salary_data)
data.head(10)
```

```python
data.replace(to_replace=['<=50k','>50k'],value=[0,1],inplace=True)
data.head(10)
```

```python
average_salary_by_workclass = data.groupby('workclass')['income'].agg('mean')
highest_salary_workclass = average_salary_by_workclass.idxmax()
highest_salary_value = average_salary_by_workclass.max()

print(f"The work class with the highest average salary is '{highest_salary_workclass}' with an average salary of {highest_salary_value:.2f}.")
```

```python
gender_salary_distribution = data.groupby('gender')['income'].apply(lambda x: (x == '>50K').sum() / len(x) * 100)

# Display the results
print(gender_salary_distribution)
```
